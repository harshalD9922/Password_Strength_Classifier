# To make it more robustness shuffle randomly

```python
import random
random.shuffle(password_tuple)
```

```python
x=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
```

```python
x
```

Create a custom function to split input character of list

```python
def word_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_char('kzde5577')
```

import TF-IDF vectorizer to convert string data into numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_char)
```

Apply TF-IDF vectorizer on data

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

```python
first_doc_vector=X[0]
first_doc_vector
```

```python
first_doc_vector.T.todense()
```

```python
data1=pd.DataFrame(first_doc_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
data1.sort_values(by=['TF-IDF'],ascending=False)
```

# split data into train & test

train---> To learn the relationship within data, 
test-->  To do predictions, and this testing data will be unseen to my model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cla=LogisticRegression(random_state=0,multi_class="multinomial")
cla.fit(X_train,y_train)
```

# prediction for specific custom dataset

```python
dt=np.array(['%@123abcd'])
pred=vectorizer.transform(dt)
cla.predict(pred)
```

# Prediction on test-dataset

```python
y_pred=cla.predict(X_test)
y_pred
```

```python
#check acciracy of our model using confusion_matrix
#and accuracy score
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```

# Create report of your Model

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
